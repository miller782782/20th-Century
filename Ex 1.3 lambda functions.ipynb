```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

```python
# Create a dataframe

data = pd.DataFrame({"Name": ['Katelyn', 'Sarah', 'Adrian', 'Heather', 'Alan', 'Jade'],
                     "Gender": ['F','F','M','F','M','NB'],
                     "Age": [36,31,45,43,40,31]})
```

```python
data
```

```python
# Use a lambda function to calculate year of birth

yob = lambda x: 2025-x
yob(data['Age'])
```

```python
# Add Year_of_Birth column to the dataframe

data['Year_of_Birth']=yob(data['Age'])

data
```

```python
# Create a flag column from the a text column using lambda and apply

# The flag column here tells whether the name has fewer than 5, exactly 5 or more than 5 letters

data['Name_length'] = data['Name'].apply(
    lambda x:'fewer than 5 letters' if len(x)<5 else ('more than 5 letters' if len(x)>5 else '5 letters'))

data
```
